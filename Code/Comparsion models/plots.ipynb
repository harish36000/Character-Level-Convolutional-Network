## Accuracies of Different Models

```python
import numpy as np
import matplotlib.pyplot as plt

# Name of Models used
Model = ['CNN','BOW','n-gram','LSTM']
#Accuracies of model observed
Acc = [0.84,0.78,0.79,0.77]

#Plotting bar graph
fig = plt.figure(figsize =(8, 5))
plt.bar(Model, Acc)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy for different models')
plt.show()
```

## Accuracies of Different dataset

```python
# Name of Models used
# Dataset 1 -> 30 percent of whole 1.6M tweets
# Dataset 2 -> 10 percent of whole
Model = ['dataset1 CNN','dataset2 CNN','dataset1 ngram TFIDF','dataset2 ngram TFIDF']
#Accuracies of model observed
Acc = [0.84,0.76,0.79,0.77]

#Plotting bar graph
fig = plt.figure(figsize =(10, 5))
plt.bar(Model, Acc)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Accuracy for different datasets')
plt.show()
```
